```
import pandas as pd
from radiomics import featureextractor
```

```
H# Import the os module, for the os.walk function import os
# Set the directory 
rootDir = '/content/drive/My Drive/509/Data'
dirName_list = []
for dirName, subdirList, fileList in os.walk(rootDir):
  dirName_list.append(dirName)
  print('Found directory: %s' % dirName) 
  for fname in fileList:
    print('\t%s' % fname)
```

```
rootDir = '/content/drive/My Drive/509/Data'
imagePath = []
labelPath = []
for dirName, subdirList, fileList in os.walk(rootDir): 
  print(fileList)
  if len(fileList) == 1:
    continue
  imagePath.append(dirName+'/'+[file for file in fileList if 'image' in file][0])
  labelPath.append(dirName+'/'+[file for file in fileList if 'gz' not in file][0])
```

```
imagePath
```

```
labelPath
```

```
for i in imagePath, labelPath:
  print(i)
```

```
# Instantiate the extractor
extractor = featureextractor.RadiomicsFeatureExtractor() #extracting features
result = []
for i in range(10):
  result.append(extractor.execute(imagePath[i], labelPath[i]))
```

```
# extractor
# result
```

```
result = pd.DataFrame(result)
result
```

```
def transform_result_to_csv(result):
  res = pd.DataFrame(result)
  ID = [i.split('/')[-1].split('_')[0] for i in dirName_list][1:]
  res['ID'] = ID
  res.iloc[:,0] = res['ID']
  res.drop('ID', axis = 1, inplace=True)
  res.insert(0,'ID',ID)
  p_result = '/content/drive/My Drive/509/result.csv'
  res.to_csv(p_result)

transform_result_to_csv(result)
```
